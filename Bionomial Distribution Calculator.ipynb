```python
import numpy as np
import scipy.special
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.style.use('ggplot')
```

```python
def binomial(n, p): #n=Number Of Trials #p=Probability
    # 0, 1, 2, ..., n
    rs = np.arange(0,n+1)
    # nCr/2**n
    ps = scipy.special.comb(n,rs)*p**rs*(1-p)**(n-rs)
    
    # Cumulative probability
    cs = np.cumsum(ps)
    
    # Critical region
    s = 0.05/2 #Two-Tailed Test, Critical Region = Areas that would Identify the Null Hypothesis to be correct or Incorrect
    cl, ch = s, 1-s #From the top region  and the bottom region
    indices = np.logical_and(cs >= cl, cs <= ch)
    
    f, ax = plt.subplots()
    f.set_figwidth(12)
    f.set_figheight(6)

    ax.set_xticks(rs)
    ax.scatter(rs, cs, c='blue',alpha=0.5)
    ax.scatter(rs[indices], ps[indices], c='green', alpha=0.5)
    ax.scatter(rs[~indices], ps[~indices], c='red', alpha=0.5)
    return ax
binomial(100, 0.7);
```
